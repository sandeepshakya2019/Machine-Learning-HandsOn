## Basic Plotting with seaborn

```python
tips_dataset = sns.load_dataset('tips')
tips_dataset.head()
```

```python
## create a scattar plot
import matplotlib.pyplot as plt

sns.scatterplot(x="total_bill", y="tip", data=tips_dataset)
plt.title("Scatter plot using Seaborn")
plt.show()
```

```python
## line plot
sns.lineplot(x="size", y="total_bill", data=tips_dataset)
plt.title("Line plot using Seaborn")
plt.show()
```

```python
## Categorical plot
sns.barplot(x="sex", y="total_bill", data=tips_dataset)
plt.title("Bar plot using Seaborn")
plt.show()
```

```python
## Categorical plot (bar pot)
sns.boxplot(x="day", y="total_bill", data=tips_dataset)
plt.title("Categorical plot using Seaborn")
plt.show()
```

```python
## Categorical plot (violin plot)
sns.violinplot(x="day", y="total_bill", data=tips_dataset)
plt.title("Categorical plot using Seaborn")
plt.show()
```

```python
## Categorical plot (histogram pot)
sns.histplot(tips_dataset['total_bill'],bins=10, kde=True )
plt.title("Categorical plot using Seaborn")
plt.show()
```

```python
## kde plot
sns.kdeplot(tips_dataset['total_bill'],fill=True)
plt.show()
```

```python
# pair plot
sns.pairplot(tips_dataset)
# just regression varibale is in consideration
```

```python
# heat map (only numerical featires)
corr = tips_dataset[['total_bill', 'size', 'tip']].corr()
# relation between one feature to another feature
corr, sns.heatmap(corr, annot=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters for the Poisson distribution
lambdas = [1, 4, 7]  # Different rate parameters (λ)
k_values = np.arange(0, 20)  # Values of k

# Plotting
plt.figure(figsize=(10, 6))

for lam in lambdas:
    pmf_values = poisson.pmf(k_values, lam)
    plt.plot(k_values, pmf_values, marker='o', label=f'λ = {lam}')

# Adding labels, legend, and grid
plt.title('Poisson Distribution for Different λ Values', fontsize=16)
plt.xlabel('Number of Events (k)', fontsize=14)
plt.ylabel('Probability (P(X = k))', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(alpha=0.5)
plt.legend(fontsize=12)
plt.show()
```

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
# Parameters for the Normal (Gaussian) distribution
means = [0, 2]  # Different means (μ)
std_devs = [1, 2]  # Different standard deviations (σ)

# Range for the x-axis
x = np.linspace(-10, 10, 500)

# Plotting
plt.figure(figsize=(10, 6))

for mean in means:
    for std_dev in std_devs:
        pdf_values = norm.pdf(x, mean, std_dev)
        plt.plot(x, pdf_values, label=f'μ = {mean}, σ = {std_dev}')

# Adding labels, legend, and grid
plt.title('Normal Distribution for Different μ and σ Values', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('Probability Density', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(alpha=0.5)
plt.legend(fontsize=12)
plt.show()
```

```python
# Plotting the Empirical Rule (68-95-99.7) for a Normal Distribution
mean = 0  # Mean (μ)
std_dev = 1  # Standard deviation (σ)

# Range for the x-axis
x = np.linspace(-4, 4, 500)
pdf_values = norm.pdf(x, mean, std_dev)

# Plot the normal curve
plt.figure(figsize=(10, 6))
plt.plot(x, pdf_values, label='Normal Distribution (μ=0, σ=1)', color='blue')

# Shade regions for 68%, 95%, and 99.7%
regions = [1, 2, 3]
colors = ['lightblue', 'lightgreen', 'lightcoral']
labels = ['68% within ±1σ', '95% within ±2σ', '99.7% within ±3σ']

for i, region in enumerate(regions):
    plt.fill_between(x, 0, pdf_values, where=(x >= -region*std_dev) & (x <= region*std_dev),
                     color=colors[i], alpha=0.4, label=labels[i])

# Adding labels, legend, and grid
plt.title('Empirical Rule for a Normal Distribution (68-95-99.7)', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('Probability Density', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.axvline(mean, color='black', linestyle='--', linewidth=1, label='Mean (μ)')
plt.grid(alpha=0.5)
plt.legend(fontsize=12)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the uniform distribution
a = 2  # lower bound
b = 6  # upper bound

# Values for the PDF
x_pdf = np.linspace(1, 7, 400)  # x values from 1 to 7
y_pdf = np.piecewise(x_pdf, [x_pdf < a, (x_pdf >= a) & (x_pdf <= b), x_pdf > b], [0, 1/(b-a), 0])

# Values for the CDF
x_cdf = np.linspace(1, 7, 400)  # x values from 1 to 7
y_cdf = np.piecewise(x_cdf, [x_cdf < a, (x_cdf >= a) & (x_cdf <= b), x_cdf > b], [0, (x_cdf - a)/(b-a), 1])

# Plot the PDF
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x_pdf, y_pdf, label="PDF", color='blue')
plt.title('Probability Density Function (PDF)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axvline(a, color='red', linestyle='--', label="a=2")
plt.axvline(b, color='green', linestyle='--', label="b=6")
plt.legend()

# Plot the CDF
plt.subplot(1, 2, 2)
plt.plot(x_cdf, y_cdf, label="CDF", color='orange')
plt.title('Cumulative Distribution Function (CDF)')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.axvline(a, color='red', linestyle='--', label="a=2")
plt.axvline(b, color='green', linestyle='--', label="b=6")
plt.legend()

plt.tight_layout()
plt.show()
```
